#### Machine Learning Project Description :

1.	Name of the Datasets : Household Power Consumption.

2.	Dataset is a minute wise data logging of Power Consumption of a Household Consumption

3.	Data Record Duration : December 2006 to November 2010 (4 Years).

4.	Record Frequency : Per minute

5.	Features recorded (07 No.):

      a.	Global Active Power - The total active power consumed by the household (kilowatts).]

      b.	Global Reactive Power - The total reactive power consumed by the household (kilowatts).]

      c.	Voltage - Average voltage (volts)]

      d.	Global Intensity - Average current intensity (amps)]

      e.	Sub metering 1 - Active energy for kitchen (watt-hours of active energy).

      f.	Sub metering 2 - Active energy for laundry (watt-hours of active energy).

      g.	Sub metering 3 - Active energy for climate control systems (watt-hours of active energy).
      

6.	Observations : 2075259 (around 2 million)
